```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("weather")
plt.plot(x,y,color='orange',linewidth=5,linestyle='dotted')
```

# format string

```python
plt.plot(x,y,'b*--',markersize=10)
```

```python
plt.plot(x,y,'b*--',alpha=0.2)
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel("days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")

plt.legend(loc="best",shadow=True,fontsize="small")
#it put legend where blank space available
plt.grid()
```

```python
import numpy as np
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
ypos=np.arange(len(company))
ypos
```

```python
plt.ylabel("Revenue(bln)")
plt.xlabel("Company")
plt.title("Company Data",)
plt.bar(ypos,revenue)
plt.bar(ypos,profit)
```

```python
plt.ylabel("Revenue(bln)")
plt.xlabel("Company")
plt.title("Company Data",)
plt.bar(ypos,revenue,label='Revenue')
plt.bar(ypos,profit,label='Profit')
plt.legend()
```

```python
plt.ylabel("Revenue(bln)")
plt.xlabel("Company")
plt.title("Company Data",)
plt.bar(ypos-0.2,revenue,width=0.4,label='Revenue')
plt.bar(ypos+0.2,profit,width=0.4,label='Profit')
plt.legend()
```

```python
plt.ylabel("Revenue(bln)")
plt.xlabel("Company")
plt.title("Company Data",)
plt.barh(ypos-0.2,revenue,label='Revenue')
plt.barh(ypos+0.2,profit,label='Profit')
plt.legend()
```

# Histogram

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist([blood_sugar],bins=[80,100,125,150],rwidth=0.80,color="green",histtype='step')
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'])
plt.legend()
```

# pie chart

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.pie(exp_vals,labels=exp_labels, shadow=True)
plt.axis("equal")
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2],counterclock=True, startangle=145)
plt.show()
```
